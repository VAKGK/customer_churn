# Import all the Requirement Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import lightgbm as lgb
from xgboost  import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,classification_report
import joblib as jb
```

# Load the Dataset

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

## Mpping the Gender with 0 and 1 

```python
df["Gender"] = df["Gender"].map({'Female': 0, 'Male': 1})
```

```python
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)
```

## Creating the Features Data 

```python
x = df.drop(columns = ["RowNumber","CustomerId","Exited","Surname"])
```

```python
x
```

## Creating Target Dataset

```python
y  =df["Exited"]
```

```python
y
```

## Find the Correlation to Every Feature

```python
corr_matrix = x.corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title("Feature-to-Feature Correlation Heatmap")
plt.show()
```

## split the Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.fit(x_test)
```

## Classify which Algorithms need Scaling and which not 

# LogisticRegression

```python
lr = LogisticRegression()
lr.fit(x_train_sc,y_train)
lr_pred = lr.predict(x_test)
lr_pred

lr_accuracy = accuracy_score(y_test, lr_pred)
lr_recall = recall_score(y_test, lr_pred)
lr_precision = precision_score(y_test, lr_pred)
lr_f1score = f1_score(y_test, lr_pred)
lr_class_report= classification_report(y_test,lr_pred)

print("Logistic Regression Model Evaluation:")
print(f"Accuracy: {lr_accuracy:.4f}")
print(f"recall_score: {lr_recall:.4f}")
print(f"precision: {lr_precision:.4f}")
print(f"f1score: {lr_f1score:.4f}")
# print("\nClassification Report:\n", lr_class_report)
```

# KNN Neighbours

```python
kn = KNeighborsClassifier(n_neighbors=15)
kn.fit(x_train_sc,y_train)
kn_pred = kn.predict(x_test)

kn_accuracy = accuracy_score(y_test, kn_pred)
kn_recall = recall_score(y_test, kn_pred)
kn_precision = precision_score(y_test, kn_pred)
kn_f1score = f1_score(y_test, kn_pred)
kn_class_report= classification_report(y_test,kn_pred)

print("KNN Model Evaluation:")
print(f"Accuracy: {lr_accuracy:.4f}")
print(f"recall_score: {lr_recall:.4f}")
print(f"precision: {lr_precision:.4f}")
print(f"f1score: {lr_f1score:.4f}")
```

# DecisionTree

```python
dc = DecisionTreeClassifier(criterion="entropy",max_depth=4,max_features="log2",splitter="best")
dc.fit(x_train,y_train)
dc_pred = dc.predict(x_test)
dc_pred

dc_accuracy = accuracy_score(y_test, dc_pred)
dc_recall = recall_score(y_test, dc_pred)
dc_precision = precision_score(y_test, dc_pred)
dc_f1score = f1_score(y_test, dc_pred)
dc_class_report= classification_report(y_test,dc_pred)

print("Decision Tree Model Evaluation:")
print(f"Accuracy: {dc_accuracy:.4f}")
print(f"recall_score: {dc_recall:.4f}")
print(f"precision: {dc_precision:.4f}")
print(f"f1score: {dc_f1score:.4f}")
```

# Naive Baye's

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
nb_pred = nb.predict(x_test)

nb_accuracy = accuracy_score(y_test, nb_pred)
nb_recall = recall_score(y_test, nb_pred)
nb_precision = precision_score(y_test, nb_pred)
nb_f1score = f1_score(y_test, nb_pred)
nb_class_report= classification_report(y_test,nb_pred)

print("Naive Baye's Model Evaluation:")
print(f"Accuracy: {nb_accuracy:.4f}")
print(f"recall_score: {nb_recall:.4f}")
print(f"precision: {nb_precision:.4f}")
print(f"f1score: {nb_f1score:.4f}")
```

# RandomForest

```python
rf = RandomForestClassifier(
    n_estimators=500,
    class_weight='balanced_subsample',
    max_depth=18,
    min_samples_split=7,
    min_samples_leaf=4,
    random_state=42
)
rf.fit(x_train,y_train)
rf_pred = rf.predict(x_test)
rf_pred

rf_accuracy = accuracy_score(y_test, rf_pred)
rf_recall = recall_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred)
rf_f1score = f1_score(y_test, rf_pred)
rf_class_report= classification_report(y_test,rf_pred)

print("Random Forest Model Evaluation:")
print(f"Accuracy: {rf_accuracy:.4f}")
print(f"recall_score: {rf_recall:.4f}")
print(f"precision: {rf_precision:.4f}")
print(f"f1score: {rf_f1score:.4f}")
```

# Super Vector 

```python
svc = SVC()
svc.fit(x_train_sc,y_train)
svc_pred = svc.predict(x_test)
svc_pred

svc_accuracy = accuracy_score(y_test, svc_pred)
svc_recall = recall_score(y_test, svc_pred)
svc_precision = precision_score(y_test, svc_pred)
svc_f1score = f1_score(y_test, svc_pred)
svc_class_report= classification_report(y_test,svc_pred)

print("Suoper Vector Model Evaluation:")
print(f"Accuracy: {svc_accuracy:.4f}")
print(f"recall_score: {svc_recall:.4f}")
print(f"precision: {svc_precision:.4f}")
print(f"f1score: {svc_f1score:.4f}")
```

# XGBoost

```python
# Calculate class imbalance ratio
ratio = sum(y_train == 0) / sum(y_train == 1)
print("Scale_pos_weight ratio:", ratio)

# Initialize model
xgb = XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=ratio,   # For imbalanced classes
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

# âœ… Fit the model
xgb.fit(x_train, y_train)

# Predict
xgb_pred = xgb.predict(x_test)

# Evaluation
xgb_accuracy = accuracy_score(y_test, xgb_pred)
xgb_recall = recall_score(y_test, xgb_pred)
xgb_precision = precision_score(y_test, xgb_pred)
xgb_f1score = f1_score(y_test, xgb_pred)

print("XGBoost Model Evaluation:")
print(f"Accuracy: {xgb_accuracy:.4f}")
print(f"Recall: {xgb_recall:.4f}")
print(f"Precision: {xgb_precision:.4f}")
print(f"F1 Score: {xgb_f1score:.4f}")
```

# LightGBM

```python
# LightGBM example
import lightgbm as lgb

lgbm = lgb.LGBMClassifier(
    n_estimators=500,
    max_depth=12,
    learning_rate=0.05,
    class_weight='balanced',
    random_state=42
)
lgbm.fit(x_train, y_train)

lgbm_pred = lgbm.predict(x_test)
lgbm_pred


lgbm_accuracy = accuracy_score(y_test, lgbm_pred)
lgbm_recall = recall_score(y_test, lgbm_pred)
lgbm_precision = precision_score(y_test, lgbm_pred)
lgbm_f1score = f1_score(y_test, lgbm_pred)
lgbm_class_report= classification_report(y_test,lgbm_pred)

print("Lightgbm Model Evaluation:")
print(f"Accuracy: {lgbm_accuracy:.4f}")
print(f"recall_score: {lgbm_recall:.4f}")
print(f"precision: {lgbm_precision:.4f}")
print(f"f1score: {lgbm_f1score:.4f}")
```

# CatBoost Model

```python

# Initialize model
cat = CatBoostClassifier(
    iterations=500,          # number of trees
    depth=6,                 # tree depth
    learning_rate=0.05,      # step size
    random_seed=42,
    eval_metric='F1',
    verbose=100,
    class_weights=None       # or provide class weights if imbalance
)

# If you have categorical features, pass their column indices
# cat_features = [0, 3, 5]  # example
# cat.fit(X_train, y_train, cat_features=cat_features)

cat.fit(x_train, y_train)
cat_pred  = cat.predict(x_test)
cat_pred

cat_accuracy = accuracy_score(y_test, cat_pred)
cat_recall = recall_score(y_test, cat_pred)
cat_precision = precision_score(y_test, cat_pred)
cat_f1score = f1_score(y_test, cat_pred)
cat_class_report= classification_report(y_test,cat_pred)

print("cat Model Evaluation:")
print(f"Accuracy: {cat_accuracy:.4f}")
print(f"recall_score: {cat_recall:.4f}")
print(f"precision: {cat_precision:.4f}")
print(f"f1score: {cat_f1score:.4f}")
```
